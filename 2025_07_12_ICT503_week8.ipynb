```python
#Covert to df
import pandas as pd
import numpy as np

students_df = pd.read_csv('students.csv')
courses_df = pd.read_csv('courses.csv')
enrollments_df = pd.read_csv('enrollments.csv')

```

```python
# 🔹 1. Add a New Column: Age in Months (students_df)


students_df['age_in_months'] = np.multiply(students_df['age'], 12)


print("Students with Age in Months:\n", students_df)
```

```python
# 🔹 2. Normalise Grades (0 to 1 range) (enrollments_df)

grades = enrollments_df['grade'].values
min_grade = np.min(grades)
max_grade = np.max(grades)

# Normalisation: (grade - min) / (max - min)
enrollments_df['grade_normalised'] = (grades - min_grade) / (max_grade - min_grade)
print("\nEnrollments with Normalised Grades:\n", enrollments_df)

```

```python

# 3. Calculate Mean and Standard Deviation of Grades
mean_grade = np.mean(enrollments_df['grade'])
std_grade = np.std(enrollments_df['grade'])

print(f"\nMean Grade: {mean_grade:.2f}")
print(f"Standard Deviation of Grades: {std_grade:.2f}")
```

```python
# 4. Sort Students by Age (Descending)
students_sorted = students_df.sort_values(by='age', ascending=False)
print("\nStudents Sorted by Age (Descending):\n", students_sorted)

```

```python

# 🔹 5. Sort Enrollments by Grade (Highest to Lowest)
# python
enrollments_sorted = enrollments_df.sort_values(by='grade', ascending=False)
print("\nEnrollments Sorted by Grade (Descending):\n", enrollments_sorted)
```

```python
# 🔹 6. Sort by Multiple Columns (Age then Name)
# python
students_sorted_multi = students_df.sort_values(by=['age', 'name'], ascending=[False, True])
print("\nStudents Sorted by Age then Name:\n", students_sorted_multi)
```
